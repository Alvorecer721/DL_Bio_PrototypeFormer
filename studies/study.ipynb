```python
import pickle
import optuna
```

```python
with open("swissprot_no_backbone_studies.pkl", 'rb') as f:
    studies = pickle.load(f)
```

```python
with open("tabula_muris_studies.pkl", 'rb') as f:
    studies = pickle.load(f)
```

```python
import matplotlib.pyplot as plt

ax = optuna.visualization.matplotlib.plot_parallel_coordinate(studies, 
                                                              params=[
                                                               #  "lr", 
                                                               #  "weight_decay", 
                                                               #  "n_sub_support",  
                                                                "n_layer", 
                                                                "n_head", 
                                                               #  "contrastive_coef", 
                                                               #  "dropout", 
                                                            #    "norm_first", 
                                                               #  "contrastive_loss", 
                                                               #  "ffn_dim"
                                                               ])

ax.set_title("Parallel Coordinate Plot: Norm Type for Swissprot")

plt.tight_layout()
```

```python
# Create a new study object with the filtered trials
filtered_trials = [trial for trial in studies.trials if (trial.params.get('contrastive_loss') == "infoNCE" and trial.values[0] > 86)]
filtered_study = optuna.create_study(directions=studies.directions)
filtered_study.add_trials(filtered_trials)

# Generate the parallel coordinate plot for the filtered study
ax = optuna.visualization.matplotlib.plot_parallel_coordinate(
    filtered_study, 
    params=["contrastive_coef", "contrastive_loss"]
)

ax.set_title("Parallel Coordinate Plot: InfoNCE", fontsize=10)

plt.tight_layout()
plt.show()  # Show the plot
```

```python
# Create a new study object with the filtered trials
filtered_trials = [trial for trial in studies.trials if (trial.params.get('contrastive_loss') == "original" )] # and trial.values[0] > 70
filtered_study = optuna.create_study(directions=studies.directions)
filtered_study.add_trials(filtered_trials)

# Generate the parallel coordinate plot for the filtered study
ax = optuna.visualization.matplotlib.plot_parallel_coordinate(
    filtered_study, 
    params=["contrastive_coef", "contrastive_loss"]
)

ax.set_title("Parallel Coordinate Plot: Original PCL", fontsize=10)

plt.tight_layout()
plt.show()  # Show the plot
```
