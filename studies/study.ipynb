```python
import pickle
import optuna
```

```python
with open("tabula_muris_studies.pkl", 'rb') as f:
    tabula_mruis_studies = pickle.load(f)
```

```python
import matplotlib.pyplot as plt

optuna.visualization.matplotlib.plot_parallel_coordinate(tabula_mruis_studies, 
                                                         params=[
                                                            #  "lr", 
                                                            #  "weight_decay", 
                                                            #  "n_sub_support",  
                                                            #  "n_layer", 
                                                            #  "n_head", 
                                                             "contrastive_coef", 
                                                            #  "dropout", 
                                                            #  "norm_first", 
                                                             "contrastive_loss", 
                                                            #  "ffn_dim"
                                                             ])
plt.tight_layout()
```

```python
# Create a new study object with the filtered trials
filtered_trials = [trial for trial in tabula_mruis_studies.trials if trial.params.get('contrastive_loss') == "original"]
filtered_study = optuna.create_study(directions=tabula_mruis_studies.directions)
filtered_study.add_trials(filtered_trials)

# Generate the parallel coordinate plot for the filtered study
optuna.visualization.matplotlib.plot_parallel_coordinate(
    filtered_study, 
    params=["contrastive_coef", "contrastive_loss"]
)

plt.tight_layout()
plt.show()  # Show the plot
```
